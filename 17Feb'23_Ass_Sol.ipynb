```python
#Q1. What is MongoDB? Explain non-relational databases in short. In which scenarios it is preferred to use MongoDB over SQL databases?
#Ans-

'''MongoDB is a popular open-source NoSQL database that stores data in a document-oriented format using JSON-like documents with dynamic schemas. 
MongoDB is designed to scale horizontally and handle large volumes of unstructured and semi-structured data, making it a popular choice for modern web and mobile applications.

Non-relational databases, also known as NoSQL databases, are databases that do not use the traditional relational data model based on tables and schema. 
Instead, they use flexible data models that can handle unstructured, semi-structured, and polymorphic data types. 
NoSQL databases are highly scalable and can handle large volumes of data and are often used in scenarios where data is rapidly changing or requires flexible and fast queries.

MongoDB is preferred over SQL databases in scenarios where there is a need for high scalability, performance, and flexibility. 
For instance, MongoDB is often used in applications that require real-time analytics, real-time data processing, and high-speed data ingestion. 
Also, it is commonly used in modern web and mobile applications that require flexible data models and schemaless databases. 
SQL databases, on the other hand, are preferred in scenarios where data is highly structured and the relationships between data are well-defined, such as in financial systems or enterprise resource planning systems.'''
```

```python
#Q2. State and Explain the features of MongoDB.
#Ans-

'''Document-Oriented: MongoDB stores data in JSON-like documents with dynamic schemas, which means it can handle data with varying fields and structures. 
This makes it ideal for handling unstructured and semi-structured data.

1. High Scalability: MongoDB is designed to scale horizontally across multiple servers and clusters, which allows it to handle large volumes of data and traffic. 
This also means that it can handle increasing workload without compromising on performance.

2. Flexible Data Model: MongoDB's flexible data model allows for easy modification and updating of data, as well as support for nested and hierarchical data structures. 
This means that developers can easily modify their data models as their application evolves.

3. Indexing and Querying: MongoDB has powerful indexing and querying capabilities, including support for text search, geospatial queries, and aggregation queries. 
This makes it easy to find and analyze data quickly and efficiently.

4. Automatic Sharding: MongoDB's automatic sharding feature distributes data across multiple servers, which allows for better load balancing and improved performance.

5. Easy Integration: MongoDB integrates easily with popular programming languages, such as Python, Java, and Node.js, and with popular tools and frameworks, such as Hadoop and Spark.

6. High Availability: MongoDB offers automatic failover and replication features, which ensure that data is always available, even in the event of hardware failure or network issues.

Overall, MongoDB's features make it a powerful and flexible database solution for modern applications that require scalability, performance, and flexibility.'''
```

```python
#Q3. Write a code to connect MongoDB to Python. Also, create a database and a collection in MongoDB.
#Ans-

'''Here is an example code to connect MongoDB to Python using the PyMongo driver and create a database and a collection:'''

# import the required libraries
import pymongo

# create a client object to connect to MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/")

# create a new database called "mydatabase"
mydb = client["mydatabase"]

# create a new collection called "customers" in the "mydatabase" database
mycol = mydb["customers"]

# add a document to the "customers" collection
mydict = { "name": "John", "address": "Highway 37" }
x = mycol.insert_one(mydict)

# print the ID of the newly inserted document
print(x.inserted_id)


'''In this code, we first import the pymongo library to connect to MongoDB. We then create a client object that connects to the local MongoDB instance running on the default port (27017).

Next, we create a new database called "mydatabase" using the client object. We also create a new collection called "customers" in this database using the mydb object.

Finally, we add a new document to the "customers" collection using a dictionary object called mydict. 
We insert this document using the insert_one method of the mycol object, which returns the ID of the newly inserted document. We then print this ID using the print function.'''
```

```python
#Q5. Explain how you can use the find() method to query the MongoDB database. Write a simple code to demonstrate this.
#Ans-

'''The find() method in MongoDB is used to query a collection and retrieve documents that match a specific set of criteria. 
It returns a cursor object that we can use to iterate over the matching documents.

The basic syntax of the find() method is as follows:'''

db.collection.find(query, projection)

'''Here, db is the name of the database, collection is the name of the collection to query, query is an optional parameter that specifies the query criteria, and projection is an optional parameter that specifies the fields to include or exclude from the result.

The query parameter is a JSON document that specifies the conditions that the documents must meet in order to be returned. 
For example, to find all documents in a collection with a specific value for a field, we can use the following query:'''

query = { "field_name": "field_value" }

'''We can also use various comparison operators like $gt, $lt, $gte, $lte, $ne, and logical operators like $and, $or, and $not to construct complex queries.

Here's an example code that demonstrates how to use the find() method to query a MongoDB collection:'''

# import the required libraries
import pymongo

# create a client object to connect to MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/")

# get a reference to the database and collection
db = client["mydatabase"]
collection = db["mycollection"]

# query the collection to find all documents
docs = collection.find()

# iterate over the cursor to print the matching documents
for doc in docs:
    print(doc)


'''In this code, we first create a client object that connects to the local MongoDB instance, and get a reference to a database and a collection. 
We then use the find() method without any query parameters to retrieve all documents in the collection, and store the result in a cursor object called docs. 
Finally, we iterate over the cursor object using a for loop, and print each document using the print() function.'''
```

```python
#Q6. Explain the sort() method. Give an example to demonstrate sorting in MongoDB.
#Ans-

'''The sort() method in MongoDB is used to sort the documents in a collection based on one or more fields. It takes one parameter, which is a document that specifies the sorting criteria. 
The keys in this document represent the fields to sort on, and their values indicate the sorting order.

The values can be 1 or pymongo.ASCENDING to indicate ascending order, or -1 or pymongo.DESCENDING to indicate descending order.

Here's an example code that demonstrates how to use the sort() method to sort a MongoDB collection based on a specific field:'''

# import the required libraries
import pymongo

# create a client object to connect to MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/")

# get a reference to the database and collection
db = client["mydatabase"]
collection = db["mycollection"]

# sort the documents based on a specific field
docs = collection.find().sort("field_name", pymongo.DESCENDING)

# iterate over the cursor to print the sorted documents
for doc in docs:
    print(doc)
    
'''In this example, we first create a client object that connects to the local MongoDB instance, and get a reference to a database and a collection. 
We then use the find() method without any query parameters to retrieve all documents in the collection, and call the sort() method on the result with the field name and the pymongo.DESCENDING constant to indicate descending order. 
The sorted result is stored in the docs cursor object.

Finally, we iterate over the cursor object using a for loop, and print each document using the print() function. 
The documents are printed in descending order of the specified field.'''
```

```python
#Q7. Explain why delete_one(), delete_many(), and drop() is used.
#Ans-

'''In MongoDB, delete_one(), delete_many(), and drop() are methods used to delete data from a collection or to drop the entire collection.

delete_one() is used to delete the first document that matches the specified filter criteria. 
If multiple documents match the criteria, only the first document encountered in the collection will be deleted.

delete_many() is used to delete all documents that match the specified filter criteria. 
This method is useful when you want to delete multiple documents that meet a specific condition.

drop() is used to drop the entire collection, i.e., delete all documents in the collection and remove the collection itself from the database. 
This method is useful when you want to completely remove a collection and all its data.

Here are some scenarios where you might use these methods:

Suppose you have a collection of customer data and you want to delete a specific customer's record. 
In this case, you can use delete_one() to remove the customer's document from the collection.

If you want to remove all customer records that have an overdue payment status, you can use delete_many() to delete all such documents from the collection.

If you want to completely remove the customer data collection from your database, you can use the drop() method to delete all the documents in the collection and remove the collection itself.

Overall, these methods are used to modify or remove data from a MongoDB collection, which is essential for maintaining data integrity and managing the database effectively.'''
```
